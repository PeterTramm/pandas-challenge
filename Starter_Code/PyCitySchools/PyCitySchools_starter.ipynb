```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
#Find all the school names
Total_Schools = len(school_data_complete["school_name"].unique())

#Calculate the total number of students
Total_Students =len(school_data_complete['Student ID'].unique())

#Calculate the total budget 
Total_budget = school_data['budget'].sum()

#Calculate the average maths scores 
Math_avg = student_data['maths_score'].mean()

#Calculate the average reading scores
Read_avg = student_data['reading_score'].mean()

#Calculate the percentage of students with a passing maths score (50 or greater)
m_pass_50 = student_data.loc[student_data['maths_score']>=50,:]
m_percent_pass_50 = len(m_pass_50)/Total_Students * 100

#Calculate the percentage of students with a passing reading score (50 or greater)
r_pass_50 = student_data.loc[student_data['reading_score']>=50,:]
r_percent_pass_50 = len(r_pass_50)/Total_Students * 100

#Calculate the percentage of students who passed maths and reading 
mr_pass_50 = student_data.loc[(student_data['maths_score']>=50) & (student_data['reading_score']>=50)]
mr_percent_pass_50 = len(mr_pass_50)/Total_Students * 100
```

```python
#Display results in a data frame
results_dict = {"Total Schools": Total_Schools,
                "Total Students": Total_Students,
                "Total Budget": Total_budget,
                "Average Maths Score": Math_avg,
                "Average Reading Score": Read_avg,
                "% Passing Maths": m_percent_pass_50,
                "% Passing Reading": r_percent_pass_50,
                "% Overall Passing": mr_percent_pass_50}
```

```python
#Creating a dataframe for the results
results_df = pd.DataFrame(results_dict,index=[0])
```

```python
#Clean the formatting using pandas method
results_df["Average Maths Score"] =  results_df["Average Maths Score"].map('{:.2f}%'.format)
results_df["Average Reading Score"] =  results_df["Average Reading Score"].map('{:.2f}%'.format)
results_df["% Passing Maths"] =  results_df["% Passing Maths"].map('{:.2f}%'.format)
results_df["% Passing Reading"] =  results_df["% Passing Reading"].map('{:.2f}%'.format)
results_df["% Overall Passing"] =  results_df["% Overall Passing"].map('{:.2f}%'.format)
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Grouping data by school name and school type
school_summary = school_data_complete.groupby(by=['school_name','type'])
school_summary = school_summary.mean(numeric_only=True)
```

```python
#Removing unecessary columns
school_summary = school_summary[["size","budget","maths_score","reading_score"]]
```

```python
#Store school names in a list 
school_name_list = sorted(school_data['school_name'])

#List to store percentage passing for each school
school_percent_pass_math = []
school_percent_pass_reading = []
school_overall_percent_passing = []
school_per_stu_budget = []
```

```python
#cycle through each name in the list 
for school in school_name_list:
    #Create a dataframe that only has that school name 
    school_name_summary = school_data_complete.loc[school_data_complete['school_name']==school]
    
    #find the percentage of students passing maths >=50
    sm_pass_50 = school_name_summary.loc[school_name_summary['maths_score']>=50,:]
    sm_percent_pass_50 = len(sm_pass_50)/len(school_name_summary) * 100
    
    #find the percentage of students passing reading >=50
    sr_pass_50 = school_name_summary.loc[school_name_summary['reading_score']>=50,:]
    sr_percent_pass_50 = len(sr_pass_50)/len(school_name_summary) * 100
    
    #Find the percentage of students passing both maths and reading
    smr_pass_50 = school_name_summary.loc[(school_name_summary['maths_score']>=50) & (school_name_summary['reading_score']>=50)]
    smr_percent_pass_50 = len(smr_pass_50)/len(school_name_summary) * 100
    #Find the per student budget 
    
    #add that into a list to be added to dataframe
    school_percent_pass_math.append(sm_percent_pass_50)
    school_percent_pass_reading.append(sr_percent_pass_50)
    school_overall_percent_passing.append(smr_percent_pass_50)
```

```python
#add list to dataframe
school_summary['% Passing Maths'] = school_percent_pass_math
school_summary['% Passing Reading'] = school_percent_pass_reading
school_summary['% Overall Passing'] = school_overall_percent_passing
```

```python
#Finding the per student budget 

#Create list to store values in for each school
budget_list = []
total_students = []

#List to hold the per student budget for each school
per_stu_budget = []
```

```python
#Get a list of the budgets
for school in school_name_list:
    #Create a dataframe that only contains rows for that school 
    school_name_summary = school_data.loc[school_data['school_name']==school]
    #Append each budget and total value for each school into their respective list
    budget_list.append(school_name_summary.iloc[0,4])
    total_students.append(school_name_summary.iloc[0,3])
```

```python
#Calculates the per_stu_budget for each school
for i in range(len(budget_list)):
    per_stu_budget.append(budget_list[i]/total_students[i])
```

```python
#Add the per_stu_budget column to the school_summary
school_summary['Per Student Budget'] = per_stu_budget
```

```python
#Reseting the index for the school_summary 
school_summary = school_summary.reset_index()
```

```python
#Cleaning up column names 
school_summary = school_summary.rename(columns= {'school_name':'School Name',
                                                'type':"School Type",
                                                'size':"Total Students",
                                                'budget':"Total Budget",
                                                'maths_score':'Average Maths Score',
                                                'reading_score':'Average Reading Score'})
```

```python
#Formating values 
school_summary['Total Students'] = school_summary['Total Students'].map('{:.0f}'.format)
school_summary['Total Budget'] = school_summary['Total Budget'].map(format_currency)
school_summary['Per Student Budget'] = school_summary['Per Student Budget'].map(format_currency)
```

```python
#Reordering the columns in the table
school_summary = school_summary[['School Name','School Type','Total Students','Total Budget','Per Student Budget',
                               'Average Maths Score','Average Reading Score','% Passing Maths','% Passing Reading',
                               '% Overall Passing']]
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Create bins and group names for per student budget
bins = [0,585,630,645,680]
group_names =['<$585','$585-630','$630-645','$645-680']
```

```python
#Converts value in Per Student Budget back to integers using locale
school_summary['Per Student Budget'] = [locale.atof(school_summary['Per Student Budget'][i].strip("$")) for i in range(len(school_summary['Per Student Budget']))]
```

```python
#Create a new column classifying the type of spending
school_summary['Spending Ranges'] = pd.cut(school_summary['Per Student Budget'],bins,labels=group_names,include_lowest=True)
```

```python
#Group the table by spending ranges
spending_summary = school_summary.groupby(by='Spending Ranges')
spending_summary = spending_summary.mean(numeric_only=True)
```

```python
#clean up the table 
#Removing uncessary columns 
spending_summary = spending_summary[['Average Maths Score','Average Reading Score','% Passing Maths','% Passing Reading',
                                     '% Overall Passing']]

#Rounding up numbers in each colum to 2 decimal places 
spending_summary['Average Maths Score'] = spending_summary['Average Maths Score'].map('{:.2f}'.format)
spending_summary['Average Reading Score'] = spending_summary['Average Reading Score'].map('{:.2f}'.format)
spending_summary['% Passing Maths'] = spending_summary['% Passing Maths'].map('{:.2f}'.format)
spending_summary['% Passing Reading'] = spending_summary['% Passing Reading'].map('{:.2f}'.format)
spending_summary['% Overall Passing'] = spending_summary['% Overall Passing'].map('{:.2f}'.format)
```
