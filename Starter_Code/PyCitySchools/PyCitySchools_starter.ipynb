## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
#Find all the school names
Total_Schools = len(school_data_complete["school_name"].unique())

#Calculate the total number of students
Total_Students =len(school_data_complete['Student ID'])

#Calculate the total budget 
Total_budget = school_data['budget'].sum()

#Calculate the average maths scores 
Math_avg = student_data['maths_score'].mean()

#Calculate the average reading scores
Read_avg = student_data['reading_score'].mean()

#calculate the percentage of students with a passing maths score (50 or greater)
m_pass_50 = student_data.loc[student_data['maths_score']>=50,:]
m_percent_pass_50 = len(m_pass_50)/Total_Students * 100

#calculate the percentage of students with a passing reading score (50 or greater)
r_pass_50 = student_data.loc[student_data['reading_score']>=50,:]
r_percent_pass_50 = len(r_pass_50)/Total_Students * 100

#Calculate the percentage of students who passed maths and reading 
mr_pass_50 = student_data.loc[(student_data['maths_score']>=50) & (student_data['reading_score']>=50)]
mr_percent_pass_50 = len(mr_pass_50)/Total_Students * 100
```

```python
#Display results in a data frame
results_dict = {"Total Schools": Total_Schools,
                "Total Students": Total_Students,
                "Total Budget": Total_budget,
                "Average Maths Score": Math_avg,
                "Average Reading Score": Read_avg,
                "% Passing Maths": m_percent_pass_50,
                "% Passing Reading": r_percent_pass_50,
                "% Overall Passing": mr_percent_pass_50}
results_df = pd.DataFrame(results_dict,index=[0])

```

```python
results_df.dtypes
```

```python
#Clean the formatting using pandas method
results_df["Average Maths Score"] =  results_df["Average Maths Score"].map('{:.2f}%'.format)
results_df["Average Reading Score"] =  results_df["Average Reading Score"].map('{:.2f}%'.format)
results_df["% Passing Maths"] =  results_df["% Passing Maths"].map('{:.2f}%'.format)
results_df["% Passing Reading"] =  results_df["% Passing Reading"].map('{:.2f}%'.format)
results_df["% Overall Passing"] =  results_df["% Overall Passing"].map('{:.2f}%'.format)
results_df
```

```python
#Grouping the data by school name then the type of school
school_summary = school_data_complete.groupby(['school_name','type']).mean(numeric_only=True)
school_summary.reset_index()
school_summary.head(10)
```

```python
#Checking columns
school_summary.columns
```

```python
#Removing unecessary columns
school_summary = school_summary[["size","budget","maths_score","reading_score"]]
school_summary.head()
```

```python
#find the percentage of students passing maths for each school
#Store school names in a list 
school_name_list = sorted(school_data['school_name'])

#List to store percentage passing for each school
school_percent_pass_math = []
school_percent_pass_reading = []
school_overall_percent_passing = []
school_per_stu_budget = []

#cycle through each name in the list 
for school in school_name_list:
    #Create a dataframe that only has that school name 
    school_name_summary = school_data_complete.loc[school_data_complete['school_name']==school]
    #find the percentage of students passing maths >=50
    sm_pass_50 = school_name_summary.loc[school_name_summary['maths_score']>=50,:]
    sm_percent_pass_50 = len(sm_pass_50)/len(school_name_summary) * 100
    
    #find the percentage of students passing reading >=50
    sr_pass_50 = school_name_summary.loc[school_name_summary['reading_score']>=50,:]
    sr_percent_pass_50 = len(sr_pass_50)/len(school_name_summary) * 100
    
    #Find the percentage of students passing both maths and reading
    smr_pass_50 = school_name_summary.loc[(school_name_summary['maths_score']>=50) & (school_name_summary['reading_score']>=50)]
    smr_percent_pass_50 = len(smr_pass_50)/len(school_name_summary) * 100
    #Find the per student budget 
    
    
    #add that into a list to be added to dataframe
    school_percent_pass_math.append(sm_percent_pass_50)
    school_percent_pass_reading.append(sr_percent_pass_50)
    school_overall_percent_passing.append(smr_percent_pass_50)
    
#add list to dataframe
school_summary['% Passing Maths'] = school_percent_pass_math
school_summary['% Passing Reading'] = school_percent_pass_reading
school_summary['% Overall Passing'] = school_overall_percent_passing
school_summary.head()
```

```python
#Find the per student budget 
school_summary.head()
budget_list = []
total_students = []
per_stu_budget = []
#Get a list of the budgets
for school in school_name_list:
    #Create a dataframe that only has that school name 
    school_name_summary = school_data.loc[school_data['school_name']==school]
    #Append each budget and total value for each school into their respective list
    budget_list.append(school_name_summary.iloc[0,4])
    total_students.append(school_name_summary.iloc[0,3])
    
#Calculates the per_stu_budget
for i in range(len(budget_list)):
    per_stu_budget.append(budget_list[i]/total_students[i])

#Add the per_stu_budget to the school_summary
school_summary['Per Student Budget'] = per_stu_budget
school_summary.head()
```

```python
#Reseting the index for the school_summary 
school_summary = school_summary.reset_index()
school_summary.head()
```

```python
#Cleaning up column names 
school_summary = school_summary.rename(columns= {'school_name':'School Name',
                                                'type':"School Type",
                                                'size':"Total Students",
                                                'budget':"Total Budget",
                                                'maths_score':'Average Maths Score',
                                                'reading_score':'Average Reading Score'})
```
